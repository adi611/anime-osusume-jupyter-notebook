# # Text Vectarization
# # Bag of words

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 13500,stop_words="english")
```

```python
tfidf.fit_transform(data["Tags"]).toarray()
```

```python
tfidf.fit_transform(data["Tags"]).toarray().shape
```

```python
vectors = tfidf.fit_transform(data["Tags"]).toarray()
```

```python
tfidf.get_feature_names()[0:10]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
def Recommender(user_input, rec_type):
    Index_of_anime = data[data["Name"] == user_input].index[0]
    Similarity_score = similarity[Index_of_anime]
    Sorted_scores = sorted(list(enumerate(Similarity_score)),reverse = True,key= lambda x: x[1])[1:]
    Recommended_Anime = []
    
    for i in Sorted_scores:
        if data.iloc[i[0]].Rank > 2000:
            continue
        if len(Recommended_Anime) > 5:
            break
        if rec_type is None:
            Recommended_Anime.append(data.iloc[i[0]].Name)
        elif data.iloc[i[0]].Type == rec_type:
            Recommended_Anime.append(data.iloc[i[0]].Name)
    return Recommended_Anime
```
